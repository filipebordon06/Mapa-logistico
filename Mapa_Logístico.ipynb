```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, widgets

# Par√¢metros fixos
x0 = 0.5
iteracoes = 100

# Define o mapa log√≠stico
def mapa(a, x):
    return a * x * (1 - x)

# Plot interativo com controle apenas do par√¢metro 'a'
@interact(
    a=widgets.FloatSlider(value=3.58,
                         min=1.0,
                         max=3.999,
                         step=0.001,
                         description='Par√¢metro a',
                         readout_format='.3f',
                         continuous_update=False)
)
def plot_mapa_logistico(a):
    # Gera as itera√ß√µes
    x = np.zeros(iteracoes + 1)
    x[0] = x0

    for i in range(iteracoes):
        x[i+1] = mapa(a, x[i])

    # Cria a figura
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

    # Configura√ß√µes do gr√°fico
    ax1.plot(range(iteracoes + 1), x, 'b-', linewidth=0.5, alpha=0.7)
    ax1.set_title(f'S√©rie temporal (a = {a:.4f})')
    ax1.set_xlabel('Itera√ß√£o')
    ax1.set_ylabel('Valor de x')
    ax1.grid(True, alpha=0.3)

    ax2.plot([a]*100, x[-100:], 'b.', markersize=1, alpha=0.5)
    ax2.set_title('Comportamento')
    ax2.set_xlabel('Parametro a')
    ax2.set_ylabel('Valores est√°veis')
    ax2.grid(True, alpha=0.3)

    # Adiciona informa√ß√µes sobre o comportamento
    if a < 3.0:
        plt.text(0.02, 0.95, "Converg√™ncia para ponto fixo", transform=plt.gca().transAxes,
                bbox=dict(facecolor='yellow', alpha=0.5))
    elif 3.0 <= a < 3.56995:
        plt.text(0.02, 0.95, "Comportamento peri√≥dico", transform=plt.gca().transAxes,
                bbox=dict(facecolor='orange', alpha=0.5))
    else:
        plt.text(0.02, 0.95, "Regi√£o ca√≥tica", transform=plt.gca().transAxes,
                bbox=dict(facecolor='red', alpha=0.5))

    plt.show()
```

```python
# Define o mapa log√≠stico
def mapa_logistico(a, x):
    return a * x * (1 - x)

# Par√¢metros para o gr√°fico de bifurca√ß√£o
a_min = 0
a_max = 4
num_a = 1000
iteracoes = 1000
transiente = 800  # N√∫mero de itera√ß√µes para descartar (transiente)

# Array de valores de a
a_values = np.linspace(a_min, a_max, num_a)

# Configura√ß√£o do gr√°fico
plt.figure(figsize=(12, 8))

# Para cada valor de a, calculamos as itera√ß√µes
for a in a_values:
    x = 0.5
    # Descartamos as primeiras itera√ß√µes (transiente)
    for _ in range(transiente):
        x = mapa_logistico(a, x)
    # Plotamos as pr√≥ximas itera√ß√µes
    for _ in range(iteracoes - transiente):
        x = mapa_logistico(a, x)
        plt.plot(a, x, ',k', alpha=0.25)

# Configura√ß√µes do gr√°fico
plt.title('Diagrama de Bifurca√ß√£o do Mapa Log√≠stico')
plt.xlabel('Par√¢metro a')
plt.ylabel('Valores est√°veis de x')
plt.grid(True, alpha=0.3)
plt.show()
```

**Estabilidade do ponto inst√°vel 1-peri√≥dico.**

```python
def mapa_controlado(a, x, k, x_hat):
    u = k * (x - x_hat)
    result = a * x * (1 - x) + u

    # Verifica se o n√∫mero √© finito (evita NaN ou Inf)
    if not np.isfinite(result):
        return x_hat  # reinicializa no ponto fixo inst√°vel

    # Clipping para evitar explos√µes: faixa segura [-2, 3]
    result = np.clip(result, -2, 3)

    return result

# Par√¢metros
a = 3.7
x_hat = 1 - 1/a  # ponto fixo inst√°vel ~0.7297
k_values = np.linspace(0, 3, 1000)
iterations = 1000
last = 100

# Gr√°fico
plt.figure(figsize=(10, 6))

for k in k_values:
    x = 0.5  # condi√ß√£o inicial

    # Itera√ß√µes para remover o transiente
    for _ in range(iterations - last):
        x = mapa_controlado(a, x, k, x_hat)
        if abs(x) > 1e3 or not np.isfinite(x):
            x = x_hat  # reinicia se explodir numericamente

    # Coleta dos √∫ltimos valores para o gr√°fico
    xs = []
    for _ in range(last):
        x = mapa_controlado(a, x, k, x_hat)
        if 0 <= x <= 1:
            xs.append(x)
        else:
            break  # ignora valores inst√°veis
    if xs:  # s√≥ plota se os dados forem v√°lidos
        plt.plot([k]*len(xs), xs, ',k', alpha=0.25)

plt.title('Diagrama de bifurca√ß√£o do mapa log√≠stico controlado')
plt.xlabel('k (coeficiente de feedback)')
plt.ylabel('x')
plt.xlim([0, 3])
plt.ylim([0, 1])
plt.grid(True)
plt.tight_layout()
plt.show()
```

**Controle ocasional do sistema.**

```python
def sistema_ocasional(a, x, k, x_hat, N, n):
    # Controle ocasional com satura√ß√£o de u
    if n % N == 0:
        u = k * (x - x_hat)
        result = a * x * (1 - x) + u
    else:
        result = a * x * (1 - x)

    # Verifica√ß√£o de valores finitos
    if not np.isfinite(result):
        return x_hat  # retorna ao ponto fixo se ocorrer overflow

    # Satura√ß√£o suave para manter x em faixa segura
    result = np.clip(result, -1, 2)  # valores al√©m disso causam instabilidade
    return result

# Par√¢metros do mapa
a = 3.7
x_hat = 1 - 1/a  # ponto fixo do mapa log√≠stico para esse a
N = 3
k_values = np.linspace(1.0, 2.2, 1000)  # faixa de valores de k
iterations = 5000
last = 300

# Gr√°fico
plt.figure(figsize=(10, 6))

for k in k_values:
    x = x_hat + np.random.uniform(-0.01, 0.01)  # pequena perturba√ß√£o inicial

    # Itera√ß√µes para estabilizar o sistema
    for n in range(iterations - last):
        x = sistema_ocasional(a, x, k, x_hat, N, n)
        if abs(x) > 1e3 or not np.isfinite(x):
            x = x_hat  # reinicia caso ocorra instabilidade

    # Coleta dos √∫ltimos valores para o diagrama
    xs = []
    for n in range(iterations - last, iterations):
        x = sistema_ocasional(a, x, k, x_hat, N, n)
        if 0 <= x <= 1:  # apenas valores dentro do intervalo
            xs.append(x)
        else:
            break  # interrompe se sair da faixa √∫til
    if xs:  # plota apenas se tiver pontos v√°lidos
        plt.plot([k] * len(xs), xs, ',k', alpha=0.15)

# Personaliza√ß√£o do gr√°fico
plt.title('Diagrama de bifurca√ß√£o do sistema ocasional com controle')
plt.xlabel('k (coeficiente de feedback)')
plt.ylabel('x')
plt.xlim([1.0, 2.2]) #conforme k aumenta o feedback fica mais forte e o sistema pode entrar em oscila√ß√µes podendo transicionar a um regime ca√≥tico
plt.ylim([0.45, 1])
plt.grid(True)
plt.tight_layout()
plt.show()
```
